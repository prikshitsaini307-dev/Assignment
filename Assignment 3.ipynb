```python
"""
6. Create a class named BankAccount with: 
o account_holder  
o balance  
Create methods: 
o deposit(amount)  
o withdraw(amount)   
Display the updated balance after each transaction. 
"""
```

```python
class BankAccount:
    account_holder=""
    balance=0
    def deposit(self,amount):
        self.balance += amount
        print("Balance:",self.balance)
    def withdraw(self,amount):
        self.balance -= amount
        print("Balance:",self.balance)
ac=BankAccount()
ac.account_holder="Raj"
ac.balance=1000
print("Accont holder:",ac.account_holder)
print("Initial balance:",ac.balance) 

ac.deposit(500)
ac.withdraw(300)
```
