```python
""""
8. Create a dictionary: 
students = { 
"Aman": 78, 
"Riya": 92, 
"Kriti": 88, 
"Rahul": 95 
} 
Write a program to: 
○ Find the student with the highest marks 
○ Find the student with the lowest marks 
○ Print only the students who scored more than 85 marks
"""
```

```python
students = { 
"Aman": 78, 
"Riya": 92, 
"Kriti": 88, 
"Rahul": 95 
}
highest_name = list(students.keys())[0]
lowest_name = list(students.keys())[0]

print("Students scoring more than 85:")

for name, marks in students.items():
    
    if marks > students[highest_name]:
        highest_name = name
    
    if marks < students[lowest_name]:
        lowest_name = name
   
    if marks > 85:
        print(name, "=", marks)

print("Highest marks:", highest_name, "=", students[highest_name])
print("Lowest marks:", lowest_name, "=", students[lowest_name])
```

```python
""""
9. Write a function: 
count_frequency(arr) 
that takes a list as input and prints the frequency of each element. 
Example: 
arr = [1, 2, 2, 3, 1, 4, 2] 
 Output: 
 
 1 -> 2 times 
2 -> 3 times 
3 -> 1 time 
4 -> 1 time 
"""
```

```python
def count_frequency(arr):
    
    for item in set(arr):
        freq=arr.count(item)
        if freq == 1:
            print(f"{item}->{freq} time")
        else:
            print(f"{item}->{freq} times")
arr = [1, 2, 2, 3, 1, 4, 2]  
count_frequency(arr)
              
```

```python
"""
10. Write a function: 
find_duplicates(arr) 
that takes a list as input and prints all duplicate elements present in 
the list. 
  Example: 
arr = [10, 20, 30, 20, 40, 10, 50, 30] 
  Output: 
  Duplicate elements are: 
10 
20 
30
"""
```

```python

def find_duplicates(arr):
    duplicates=[]
    for item in arr:
        if arr.count(item)>1 and item not in duplicates:
             duplicates.append(item)
    print("Duplicates elements are:")
    for item in duplicates:
        print(item)
arr=[10,20,30,20,40,10,50,30]                 
find_duplicates(arr)
```
